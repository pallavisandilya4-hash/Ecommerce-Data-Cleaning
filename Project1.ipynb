#### Dataset Column Description

This dataset contains information about customer orders placed in an e-commerce platform. Below is a brief description of each column.

**OrderID** - Unique identifier assigned to each customer order. 

**Date** - The date on which the order was placed. 

**CustomerID** - Unique identifier for each customer.

**Product** - Name of the product purchased by the customer. 

**Quantity** - Number of units of the product ordered.

**UnitPrice** - Price of a single unit of the product.

**ShippingAddress** - Delivery address where the order is shipped.

**PaymentMethod** - Payment mode used by the customer (e.g., Credit Card, UPI, PayPal). 

**OrderStatus** - Current status of the order (e.g., Delivered, Shipped, Pending, Cancelled).

**TrackingNumber** - Unique tracking number used to monitor shipment status. 

**ItemsInCart** - Total number of items present in the customer's cart at the time of purchase.

**CouponCode** - Promotional or discount coupon applied to the order, if any. 

**ReferralSource** - Source through which the customer reached the platform (e.g., Social Media, Search Engine, Direct). 

**TotalPrice** - Total amount paid for the order after considering quantity and any applicable discounts.

## Data Cleaning Summary

The following data cleaning steps were performed to prepare the dataset for analysis:

- Identified missing values using `isnull().sum()`.
- Replaced missing values in the **CouponCode** column with **"No Coupon"**, as not all customers use discount coupons.
- Checked for duplicate records using `duplicated()` and found **no duplicate rows** in the dataset.
- Converted the **Date** column from **object** to **datetime** format to enable accurate date-based analysis.
- Verified numeric columns (**Quantity, UnitPrice, ItemsInCart, and TotalPrice**) and confirmed that no invalid (zero or negative) values were present.
- The dataset is now clean, consistent, and ready for exploratory data analysis (EDA) and visualization.
